### Stack    

Stack is collection of objects         
LIFO (Last-in First Out)    
Push and Pop     
Example - a packet of tennis balls closed from the bottom.      


Application of stacks         
web brower history   
Unde operations in editing applications      
HTML docment tags matching          
evaluating arithmetic expression        
Parentheses matching       
Infixt to postfix operation     


Stacks abstract data type            
it has methods such as          
push(object): insert element              
pop(): remove and return element              
top(): returns last inserted element            
len(): returns number of elements             
isEmpty(): to check if a stack is empty   


Implementing stacks abstract data type using      
Arrays    
Linked List 

Stacks using arrays            
Built-in sequence classes       
List     


```python
class StackArray:
    """A class to define stack arrays in python"""
    def __init__(self):
        self.data = []
        self.size = 0
    
    #method to return length of stack    
    def __len__(self):
        return self.size
    
    #method to check if stack is empty or not
    def isempty(self):
        return self.size == 0
    
    #method to add an element to stack
    def push(self, e):
        self.size += 1 
        return self.data.append(e)
    
    #method to remove and return an element 
    def pop(self):
        if self.data is not self.isempty():
            self.size -= 1
            return self.data.pop()
        else: 
            return 'List is empty'   
    
    #method to return last element from the list 
    def top(self):
        if self.data is not self.isempty():
            return self.data[-1]
        else: 
            return 'List is empty'
```

```python
s.push(2)
s.push(3)
s.push(4)
s.data
```

#### Implementing Stacks using Linked list                  

We can use insert operations to insert an element     
Remove operations to remove an element at tail    


```python
class Node:
    """Class to create a node"""
    #The special attribute __slots__ allows you to explicitly state which instance attributes you 
    #expect your object instances to have, with the expected results: faster attribute access and space
    #savings in memory.
    #This is done by storing value references in __slots__ instead of __dict__
    #Denying __dict__ and __weakref__ creation if parent classes deny them and you declare __slots__
    __slot__ = 'element', '_next'
    
    #create a node with two attributes - element as value and _next as memory address for next element
    def __init__(self, element, _next):
        self.element = element
        self._next = _next
```

```python
class StacksLinked:
    """Class to implement stacks using a linked list"""
    def __init__(self):
        self._top = None #reference of head element
        self.size = 0
        
        
    #method to return length of stack
    def __len__(self):
        return self.size
    
    #method to check if a list is emplty
    def isempty(self):
        return self.size == 0
    
    #method to push an element e
    def push(self, e):
        new_node = Node(e, None)
        if self.isempty():
            self._top = new_node
        else:
            new_node._next = self._top
            self._top = new_node
        self.size += 1
        
    #method to remove and return top element from stack
    def pop(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            element = self._top.element
            self._top = self._top._next 
            self.size -= 1
            return element
    
    #method to return top element from stack
    def top(self):
        if self.isempty():
            return 'Stack is empty'
        else:
            return self._top.element
        
    #method to display elements of linked list 
    def display(self):
        p = self._top
        while p:
            print(p.element, end='  ')
            p = p._next
        
     
```

```python
s = StacksLinked()
```

```python
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.display()
```

```python
print(s.pop())
print(s.top())
```

#### Queues abstract data type          
enqueue(object): insert element

len(): returns number of elements      
isEmpty(): wheter queue is empty or not     


#### Implementing queues using arrays

```python
class QueueList:
    """class to implement queues using list"""
    def __init__(self):
        self.data = []
        self.size = 0
        
    #method to define length of queue
    def __len__(self):
        return self.size
    
    #method to check if queue is empty or not
    def isempty(self):
        return self.size == 0
    
    #method to insert element 
    def enqueue(self, e):
        self.data.append(e)
        self.size += 1
    
    #method to remove and return element
    def dequeue(self):
        if self.isempty():
            return "Queue is empty"
        else:
            self.size -= 1
            return self.data.pop(0)         
            
    #method to return first element
    def first(self):
        if self.isempty():
            return 'Queue is empty'
        else:
            return self.data[0]
```

```python
q = QueueList()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
q.data
```

```python
q.dequeue()
print(q.first())
print(len(q))
```

#### Double ended queues data structures       
Double ended queues are also collection of objects        
Supports insertion and deletion at both the front and rear end of the queue      
Waiting list for restaurant        
addfirst(object): insert elemet a the front    
addlast(object): insert element at rear 
removefirst(): remove element form the front and return element      
removelast: remove element from the last and return element         
first(): returns the element at the front           
last(): returns the element at the rear         
len(): returns number of elements        


#### Implementing DEQues using arrays 

```python
class DeQueueArray:
    def __init__(self):
        self.data = []
        self.size = 0 
    
    def __len__(self):
        return self.size
    
    def isempty(self):
        return self.size == 0
    
    def addfirst(self, e):
        self.size += 1
        return self.data.insert(0, e)
    
    def addlast(self, e):
        self.size += 1
        return self.data.append(e)
    
    def removefirst(self):
        if self.isempty():
            return 'is empty'
        else:
            return self.data.pop(0)
            self.size -= 1
    
    def removelast(self):
        if self.isempty():
            return 'is empty'
        else:
            return self.data.pop()
            self.size -= 1
            
    def first(self):
        if self.isempty():
            return 'is empty'
        else:
            return self.data[0]
        
    def last(self):
        if self.isempty():
            return 'is empty'
        else:
            return self.data[-1]
```
