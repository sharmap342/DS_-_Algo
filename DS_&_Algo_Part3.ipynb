### Tree Data Structure       

#### Trees Defination and Properties          
Tree is a non linear data structure      
In linear data stucture there is a before and after relationship, for example an element is before or after a certain element       
Non linear data structure the relationship is below or above that is some objects are below something and some objects are above something        
Tree data structure allows us to implement algorithm much faster than linear data structures      
So a tree is an abstract data structure which stores elements hierarically     
1. Set of Node
2. Nodes have parent child relationship through edges      
For n number of nodes in a tree there would be n-1 number of edges        
Tree can be empty - if it does not have any nodes      

      


#### Trees - Terminology         
Root - Which has no parent, it's on the top level     
Siblings - Children of same parent       
Leaf or External Node - Nodes which do not have any children       
Internal node -   If a node has one or more children          
Edge - An edge of tree is a pair of nodes (u,v) such that u is parent of v         
Path - Sequence of nodes such that two consecutive nodes forms an edge  for example path between p,q,a,l,e        


Subtree - Any node of tree with it's children is considered as subtree. Leaf nodes are also subrees     
Forest - A forest is a collection of trees. If you remove the root node the tree will be divided into multiple trees.       

Heigth and Levels - Heigth and levles start from root node with values 1 and 0, respectively. For level n there are n nodes in the path and n-1 edges from the bottom most node to the root node.  
The relation between hight and levels is - Height = level-1       

Degree of Node - Number of children a node has    
so degree of leaf node is 0,      

Degree of a tree - maximum of the degree of node

Binary Tree Traversals - Method of visiting all the elements of a tree                  
4 techniques for traversals     
1. Inorder Traversal          
2. Preorder Traversal             
3. Postorder Traversal          
4. Level order Traversal           

##### Preorder Traversal                  
. visit root         
. visit left subtree recursiverly preorder           
. visit right order recursively preorder         

          


##### Inorder Traversal            
. visit left subtree recursively inorder         
. visit root (root can be of tree or subtree)                     
. visit right subtree recursiverly inorder                

             


##### Postorder Traversal         
. visit left subtree recursiverly postorder
. visit right subtree recursively postorder   
. visit root      

       

##### Leverl order traversal 
. visit nodes level by level from top to bottom             
. within level visit nodes from left to right           

      

#### Easy way to remember binary trees traversals       
. Pre-oreder traversal:    +AB   prefix            
. In-order traversal:       A+B  infix           
. Post-order traversal      AB+ postfix        


### Creating binary trees          
##### Binary tree using link repersentation         


```python
class Node:
    """A class to create node for binary tree"""
    __slots__ = 'element', 'left', 'right'
    
    def __init__(self, element, left=None, right=None):
        self.element = element 
        self.left = left #reference to left child
        self.right = right #reference to right child 
```

```python
class BinaryTree:
    """A class to create binary tree"""
    def __init__(self):
        self.root = None #initiate a tree with one root node
        
    #method to create a tree 
    def make_tree(self, e, left_stree, right_stree):
        """Here left and right are objects defining left and right subtree
        - each subtree will have a root and maxium of two children"""
        self.root = Node(e, left_stree.root, right_stree.root)
        
    #method for inorder traversing
    def inorder(self, tree_root):
        if tree_root:
            self.inorder(tree_root.left)
            print(tree_root.element, end='  ')
            self.inorder(tree_root.right)
            
    #method for postorder traversing
    def postorder(self, tree_root):
        if tree_root:
            self.postorder(tree_root.left)
            self.postorder(tree_root.right)
            print(tree_root.element, end='  ')
            
    #method for preorder traversing
    def preorder(self, tree_root):
        if tree_root:
            print(tree_root.element, end='  ')
            self.preorder(tree_root.left)
            self.preorder(tree_root.right)
            
    #method for level order traversing
    levelorder = levelorder
    
    #method to calculate number of nodes in the tree 
    count_node = count_node
    
    #method to calculate height of binary tree
    count_height = count_height
```

##### Binary search tree insertion             
  

```python
# iterative insert funciotn
def insert_it(self, troot, key):
    temp = None #tmeporary node to go until the end of the tree 
    while troot:
        temp = troot
        if key == temp.element:
            return
        elif key < temp.element:
            troot = troot.left
        elif key > temp.element:
            troot = troot.right
            
    n = Node(key, None, None) #create a child node 
    if self.root:             #check if tree is empty 
        if key < temp.element:
            temp.left = n      #assign new node as a left child 
        else:
            temp.right = n     #assign new node as a right child 
    else:
        self.root = n       #if tree is empty create a root node 
```

```python
#recursive insert function   
def insert_rec(self, troot, key):
    if troot:
        if key < troot.element:
            troot.left = self.insert_rec(troot.left, key)
        if key > troot.element:
            troot.right = self.insert_rec(troot.right, key)
    else:
        n = Node(key, None, None)
        troot = n
    return troot
```
